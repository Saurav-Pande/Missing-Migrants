# Causes of Death

```python
migrants["Cause of Death"].describe()
```

```python
migrants["Cause of Death"]=migrants["Cause of Death"].str.lower()
migrants["Cause of Death"].unique()
```

```python
def compute_clustered_deaths (migrants): 
    
    migrants.loc[migrants["Cause of Death"].str.contains("starvation|malnutrition"),
                        "Cause of Death"] = "Starvation"
    migrants.loc[migrants["Cause of Death"].str.contains("murdered|Violence|shot|stabbed|excessive phyiscal abuse")
                       ,"Cause of Death"] = "Violence"
    migrants.loc[migrants["Cause of Death"].str.contains("sexual abuse|rape|attacked by apache helicopter|killed by mortar shell"),
                       "Cause of Death"] = "Violence"
    migrants.loc[migrants["Cause of Death"].str.contains("violence while attempting to board boat|air strike|killed by landmine blast|killed by mortar shell"),
                       "Cause of Death"] = "Violence"
    migrants.loc[migrants["Cause of Death"].str.contains("artillery shells|violence"),
                       "Cause of Death"] = "Violence"
    migrants.loc[migrants["Cause of Death"].str.contains("crushed|drowning|asphyxiation|suffocation"), "Cause of Death"] = "Suffocation"
    migrants.loc[migrants["Cause of Death"].str.contains("hypothermia|hyperthermia|heat stroke|harsh conditions"), "Cause of Death"] = "Harsh Conditions"
    migrants.loc[migrants["Cause of Death"].str.contains("lightning|exposure|harsh weather|exhaustion"),"Cause of Death"] = "Harsh Conditions"
    migrants.loc[migrants["Cause of Death"].str.contains("unknown|mixed|gassed|rockslide"), "Cause of Death"] = "Others"
    migrants.loc[migrants["Cause of Death"].str.contains("vehicle|fall|fire|collision|train|bus|electrocution|hit by car|fell from truck|hit by truck|plane "), 
                        "Cause of Death"] = "Accident"
    migrants.loc[migrants["Cause of Death"].str.contains("bacterial infections|head injury|cardiac arrest|pneumonia|organ failure|sickness|hypoglycemia|post-partum complications"),
                       "Cause of Death"] = "Health Conditions"
    migrants.loc[migrants["Cause of Death"].str.contains("seizure|respiratory illness|pulmonary edema|cervical cancer|heart attack|coronary artery atherosclerosis"), 
                       "Cause of Death"] = "Health Conditions"
    migrants.loc[migrants["Cause of Death"].str.contains("dehydration|dehyration"),
                       "Cause of Death"] = "Dehyrdation"
    migrants.loc[migrants["Cause of Death"].str.contains("crocodile|hippopotamus|hippoptamus|envenomation"),
                       "Cause of Death"] = "Killed by animals"
    migrants.loc[migrants["Cause of Death"].str.contains("suicide|hanging"),
                       "Cause of Death"] = "Mental Health"
    migrants.loc[migrants["Cause of Death"].str.contains("lack of access to medicine"),
                       "Cause of Death"] = "Lack of access to medicine"
    migrants.loc[migrants["Cause of Death"].str.contains("fuel burns|burned"),
                       "Cause of Death"] = "Burns"
    
    return migrants
```

```python
clustered_death=compute_clustered_deaths(migrants)
clustered_death["Cause of Death"].value_counts()
```

```python
sns.countplot(x="Cause of Death", data=migrants, order=migrants["Cause of Death"].value_counts().index)
plt.title("Causes of death recorded", size = 14)
plt.xticks(rotation=90)
plt.show()
```
